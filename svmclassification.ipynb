```python
import pandas as pd
```

```python
data = pd.read_csv("Social_Network_Ads.csv")
```

```python
data
```

```python
data = pd.get_dummies(data, dtype=int,drop_first=True)
```

```python
data
```

```python

# Axis =  1 --> Column wise delete
# Axis = 0 ---> Row wise delete
data = data.drop("User ID",axis = 1) 
```

```python
data
```

```python
Independent = data[['Gender_Male','Age','EstimatedSalary']]
```

```python
Dependent = data[['Purchased']]
```

```python
Independent
```

```python
Dependent
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Independent, Dependent, test_size = 0.30, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.svm import SVC
```

```python
classified=SVC(kernel='sigmoid',random_state=42)
```

```python
classified
```

```python
classified.fit(X_train, y_train)
```

```python
y_pred = classified.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
Clf_report = classification_report(y_test, y_pred)
```

```python
Clf_report
```

```python
Gender = int(input("Enter you gender (Male: 1, Female: 0): "))
Age = int(input("Enter you Age: "))
Estimated_Salary = int(input("Enter your Estimated Salary: "))

classified.predict([[Gender, Age, Estimated_Salary]])
```

```python
import pickle
```

```python
#save the model name and create a file need to end in.sav
fname="final_model_pred.pkl"
```

```python
# Save the model into the model name,dump=open,wb=write binary
pickle.dump(classified,open(fname,'wb'))
```
